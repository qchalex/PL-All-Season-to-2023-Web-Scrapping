```python
html = driver.page_source
soup = bs(html,'html.parser')
print(soup)
```

```python
table_headers = soup.find('thead')
print(table_headers)
```

```python
for word in table_headers.find_all('tr'):
    ray = word.find_all('th')
    for i in ray:
        print(i)
```

```python
columns = []

for word in table_headers.find_all('tr'):
    club_element = word.find('th', class_='league-table__team-header')
    
    if club_element is not None:
        club = club_element.text
        columns.append(club)

print(columns)
```

```python
for row in table_headers.find_all('tr'):
    for abbr in row.find_all('abbr'):
        columns.append(abbr.attrs['title'])

print(columns)
```

```python
for word in table_headers.find_all('tr'):
        for content in word.find_all('div',class_='thFull'):
            data = content.text
            columns.append(data)
            
print(columns)
```

```python
css_selector = '#mainContent > div.tabbedContent > div.league-table.mainTableTab.active'
element = driver.find_element(By.CSS_SELECTOR, css_selector)

element_html = element.get_attribute('outerHTML')

full_data = BeautifulSoup(element_html, 'html.parser')
```

```python
Club_names = []

team_elements = soup.find_all('span', class_='teamName')

for team in team_elements:
    team_name = team.text.strip() 
    Club_names.append(team_name)

print(Club_names)
```

```python
Club_names = []
Games_played = []
Won = []
Drawn = []
Lost = []
GF = []
GA = []
GD = []
Points = []

rows = soup.find_all('tr', attrs={'data-filtered-entry-size': '51'})

for position, row in enumerate(rows, start=1):
    columns = row.find_all('td')

    club_name = columns[1].find('span', class_='league-table__team-name').text
    Club_names.append(club_name)

    Games_played.append(columns[2].text)
    Won.append(columns[3].text)
    Drawn.append(columns[4].text)
    Lost.append(columns[5].text)
    GF.append(columns[6].text)
    GA.append(columns[7].text)
    GD.append(columns[8].text)
    Points.append(columns[9].text)
```

```python
print(len(Club_names))
print(len(Won))
print(len(Games_played))
print(len(Drawn)) 
print(len(Lost))
print(len(GF))
print(len(GA))
print(len(GD))
print(len(Points))
```

```python
data = {
    'Position': list(range(1, len(rows) + 1)),
    'Clubs': Club_names,
    'Games_Played': Games_played,
    'Game_Won': Won,
    'Games_Drawn': Drawn,
    'Games_Lost': Lost,
    'GF': GF,
    'GA': GA,
    'GD': GD,
    'Point': Points
}

pl_table = pd.DataFrame(data)

pl_table.set_index('Position', inplace=True)

print(pl_table)
```

```python
pl_table.to_csv('pl_allseason.csv', index=False)
```
